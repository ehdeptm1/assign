```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
%matplotlib inline


df = pd.read_csv('/Users/ehdep/Documents/population.csv')

list=[]

df.head()
male=df.groupby(['By administrative divisions(eup, myeon, dong)','Male (Person)'])[['Korean - male (Person)','Foreigner-Male (Person)']].mean()
female=df.groupby(['By administrative divisions(eup, myeon, dong)','Female (Person)'])[['Korean - Female (Person)','Foreigner-Female (Person)']].mean()

pd.DataFrame(male).to_csv('C:/Users/ehdep/Documents/male.csv')
pd.DataFrame(female).to_csv('C:/Users/ehdep/Documents/female.csv')

ax = male.plot(kind='bar',title='pop',figsize=(12,4),legend=True, fontsize=12)
ax.set_xlabel('By administrative divisions(eup, myeon, dong)',fontsize=12)
ax.set_ylabel('Male (Person)',fontsize=12)
ax.legend(['Korean - male (Person)','Foreigner-Male (Person)'], fontsize=12)

ax = female.plot(kind='bar',title='pop',figsize=(12,4),legend=True, fontsize=12)
ax.set_xlabel('By administrative divisions(eup, myeon, dong)',fontsize=12)
ax.set_ylabel('Female (Person)',fontsize=12)
ax.legend(['Korean - Female (Person)','Foreigner-Female (Person)'], fontsize=12)
```
